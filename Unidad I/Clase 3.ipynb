## Imputar valores NaN

```python
# Como se mostro en el codigo data['sexo'].value_counts(), el valor que mas se repite
# es de 'Mujer', por lo tanto imputaremos mediante ese valor
data['sexo'].fillna('Mujer', inplace=True)
```

```python
# Imputar los valores de la columna 'edad' mediante el promedio
data['edad'] = data["edad"].fillna(data["edad"].mean())
```

```python
# Imputar los valores de la columna 'ocupación'
data['ocupación'].value_counts()
```

```python
# Imputamos los valores de la columna 'ocupación' mediante el valor que mas se repite
# Esto solo se hace dependiendo del contexo del problema, hay casos en los que es mejor eliminar las 
# columnas que no contengan informacion
data['ocupación'].fillna('Empleado', inplace=True)
```

```python
# Imputamos los valores de la columna 'nivel_educacional' por el valor que mas se repite
data['nivel_educacional'].value_counts()
```

```python
data['nivel_educacional'].fillna('basica', inplace=True)
```

```python
# Imputar los valores de la columna 'nivel_socioeconomico'
data['nivel_socioeconomico'].value_counts()
```

```python
data['nivel_socioeconomico'].fillna('bajo', inplace=True)
```

```python
# Imputar valores de la columna sueldo mediante el promedio
data['sueldo'] = data["sueldo"].fillna(data["sueldo"].mean())
```

```python
# Verificamos las columnas NaN
data.isnull().sum()
```

## Tratamiento de Outliers

```python
# Con un boxplot podemos visualizar los valores atipicos y outliers
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.boxplot(data['edad'].dropna()) # el dropna sirve para eliminar las columnas NaN
plt.title('Boxplot de Edad')
plt.show()
```

```python
# Eliminar los outliers mediante el metodo de recorrido intercuartil

# Primero definimos los cuartiles
Q1 = data['edad'].quantile(0.25) # 1er cuartil 25%
Q3 = data['edad'].quantile(0.75) # 3er cuartil 75%
IQR = Q3 - Q1

# Luego definimos los limites mediante la 'regla de Tukey'
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar los datos para excluir outliers
data = data[(data['edad'] >= limite_inferior) & (data['edad'] <= limite_superior)]
```

```python
# Verificamos la nueva dimension del dataframe
data.shape
```

```python
# Verificamos los outliers de la columna sueldo
plt.figure(figsize=(10,6))
plt.boxplot(data['sueldo'])
plt.title('Boxplot de Edad')
plt.show()
```

```python
# podemos verificar outliers mediante un histograma
plt.figure(figsize=(10,6))
plt.hist(data['sueldo'])
plt.title('Boxplot de Edad')
plt.show()
```

```python
# Eliminar los outliers mediante el metodo de recorrido intercuartil

# Primero definimos los cuartiles
Q1 = data['sueldo'].quantile(0.25) # 1er cuartil 25%
Q3 = data['sueldo'].quantile(0.75) # 3er cuartil 75%
IQR = Q3 - Q1

# Luego definimos los limites mediante la 'regla de Tukey'
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar los datos para excluir outliers
data = data[(data['sueldo'] >= limite_inferior) & (data['sueldo'] <= limite_superior)]
```

## Visualizacion de datos con Matplotlib

```python
# Histograma de frecuencias
plt.figure(figsize=(8,4))
plt.hist(data['edad'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Grafico de barras
sexo = data['sexo'].value_counts()
plt.bar(sexo.index, sexo.values, color=['blue', 'pink'])
plt.title('Distribución por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Grafico de torta 
niveles = data['nivel_educacional'].value_counts()
plt.pie(niveles, labels=niveles.index, autopct='%1.1f%%')
plt.title('Distribución del Nivel Educativo')
plt.show()
```

```python
import seaborn as sns
sns.histplot(data=data, x='edad', kde=True)
plt.title('Distribución de Edad')
plt.show()
```

```python
# El pairplot sirve para tener una vision de la estructura de los datos que contiene el dataframe
sns.pairplot(data.dropna(), vars=['edad', 'sueldo'])
plt.show()
```

```python
# Primero, podríamos agrupar los datos por 'edad' y calcular el sueldo promedio para cada edad
# Esto es solo un ejemplo y puede no ser lo ideal dependiendo de tus necesidades de análisis
sueldo_promedio_por_edad = data.groupby('edad')['sueldo'].mean().reset_index()
sns.lineplot(x='edad', y='sueldo', data=sueldo_promedio_por_edad)
plt.title('Sueldo Promedio en Función de la Edad')
plt.xlabel('Edad')
plt.ylabel('Sueldo Promedio')
plt.show()
```
